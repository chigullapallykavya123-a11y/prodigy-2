```python
# 1. Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load the dataset
# Replace 'train.csv' with your specific file path if needed
df = pd.read_csv('train.csv')

# 3. Data Cleaning
# Display basic information
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Example: Fill missing numerical values with the mean
# df['Age'] = df['Age'].fillna(df['Age'].mean())

# Example: Drop columns with too many missing values
# df = df.drop(columns=['Cabin'])

# 4. Exploratory Data Analysis (EDA)
# Visualize the distribution of a numerical variable
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Distribution of Age')
plt.show()

# Visualize relationships (e.g., Target vs Categorical)
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df)
plt.title('Count of Survival')
plt.show()

# Correlation Heatmap for numerical variables
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
